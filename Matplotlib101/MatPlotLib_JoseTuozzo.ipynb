## 1) Data Used in This Lab

I will work with small, clean datasets to highlight each chart type:

- **Bar chart:** Average weekly study hours per subject.  
- **Line plot:** Daily temperature across a week.  
- **Pie chart:** Favorite fruit distribution in a class survey.  
- **Scatter plot:** Hours studied vs. test score (relationship).  
- **Box-and-whisker:** Score distributions for three different classes (spread, median, outliers).

Creating them below to keep the notebook self-contained.


```python
# Bar chart data: categorical (subjects) vs numerical (hours)
subjects = ["Math", "Science", "History", "English", "Art"]
avg_hours = [10, 8, 6, 7, 4]

# Line plot data: trend over time (one week of temperatures)
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temps_f = [72, 74, 70, 68, 71, 75, 77]

# Pie chart data: simple proportions
fruits = ["Apple", "Banana", "Orange", "Grapes"]
fruit_counts = [30, 25, 20, 25]

# Scatter plot data: relationship between study time and score
study_hours = np.array([2, 3, 4, 5, 6, 7, 8, 9])
scores = np.array([60, 65, 67, 72, 78, 82, 88, 93])

# Boxplot data: three classes with different distributions
class_A = [78, 85, 90, 95, 88, 76, 92, 89]
class_B = [65, 70, 68, 72, 74, 69, 71, 73]
class_C = [90, 93, 88, 96, 92, 94, 91, 89]

class_data = [class_A, class_B, class_C]

print('Datasets prepared.')
```

## 2) Bar Chart â€” *Comparing categories*

**Why this chart?** Bar charts compare **categories** against a **numerical** value.  
**Question answered:** *Which subject requires the most study time?*

**Interpretation plan:** Look for the tallest bar to identify the subject with the highest weekly study hours.


```python
plt.figure()
plt.bar(subjects, avg_hours)  # no explicit colors per lab instructions
plt.title("Average Weekly Study Hours per Subject")
plt.xlabel("Subject")
plt.ylabel("Hours per Week")
plt.tight_layout()
plt.show()

# Interpretation (printed as text for the grader):
print("Interpretation: Math shows the highest weekly study time; Art is the lowest. "
      "A bar chart is appropriate because it compares discrete categories.")
```

**This part builds a bar chart:**

*  On the x-axis, I place the subjects, and on the y-axis, I put the average hours of study.
The chart helps me compare the time I spend on each subject at a glance.

 * The taller the bar, the more time I dedicate. Shorter bars indicate those that require less effort.

*  I also included a title and axis labels to make the graph easier to read.

## 3) Line Plot â€” *Trends over time*

**Why this chart?** Line plots are ideal for **temporal trends**.  
**Question answered:** *How does temperature change throughout the week?*

**Interpretation plan:** Identify increasing/decreasing segments and any sharp changes.


```python
plt.figure()
plt.plot(days, temps_f, marker='o')  # keep defaults for style; marker helps visibility
plt.title("Daily Temperature Over a Week")
plt.xlabel("Day")
plt.ylabel("Temperature (Â°F)")
plt.grid(True)  # grid improves readability
plt.tight_layout()
plt.show()

print("Interpretation: Temperatures dip mid-week (Thu) and rise into the weekend, "
      "showing a mild upward trend by Sun. A line plot highlights these day-to-day changes.")
```

**This section creates a line plot**.
*  It takes the temperature of each day and connects them so that I can follow the change across the week.

* The purpose is to highlight the trend, not just the individual numbers.

 * Adding dots on the line makes the daily values clearer, and the grid gives the graph some structure.

*  The title and labels tell me what Iâ€™m looking at and that the data is in Fahrenheit.

## 4) Pie Chart â€” *Parts of a whole*

**Why this chart?** Pie charts show **proportions** relative to a whole.  
**Question answered:** *What share does each fruit preference occupy?*

**Interpretation plan:** Read the percentage labels to compare segments.


```python
plt.figure()
plt.pie(fruit_counts, labels=fruits, autopct="%1.1f%%", startangle=140)
plt.title("Favorite Fruit Distribution")
plt.axis('equal')  # equal aspect ratio ensures the pie is drawn as a circle.
plt.tight_layout()
plt.show()

print("Interpretation: Apple is the largest single preference; Banana and Grapes tie; "
      "Orange has a smaller but notable share. Pie is appropriate for proportion of a whole.")
```

**Pie Chart**

* This part of the code makes a pie chart.
 It uses the data about fruit choices and shows them as slices of a circle.

* The percentages appear automatically with autopct, making it easy to see the exact values.
 The idea is to understand how the whole is composed of its parts.

 * Itâ€™s a quick way to compare things like survey results or how a budget is split.

## 5) Scatter Plot â€” *Relationship between two variables*

**Why this chart?** Scatter plots reveal **relationships/correlations**.  
**Question answered:** *Do more study hours generally lead to higher scores?*

**Interpretation plan:** Look for an upward or downward trend in the cloud of points; compute correlation.


```python
plt.figure()
plt.scatter(study_hours, scores)
plt.title("Hours Studied vs. Test Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Test Score")
plt.tight_layout()
plt.show()

# Simple correlation (Pearson) for quantitative support
corr = np.corrcoef(study_hours, scores)[0,1]
print(f"Correlation (Pearson): {corr:.3f}")
print("Interpretation: There's a clear positive relationship; higher study hours associate with higher scores. "
      "A scatter plot is appropriate to visualize this relationship.")
```

**Here, the code draws a scatter chart.**

*  It maps study hours against exam results, so every dot represents a studentâ€™s data.

* The purpose is to determine the relationship between these two things.
 If the dots go upward, it suggests that more study time leads to higher marks.

 * This plot makes it easier to spot patterns, groups, or exceptions, and I also calculate the correlation to confirm what the graph shows.

## 6) Box-and-Whisker Plot â€” *Distribution, spread, and potential outliers*

**Why this chart?** Box plots summarize **median**, **quartiles**, **range**, and potential **outliers**.  
**Question answered:** *How do score distributions differ across three classes?*

**Interpretation plan:** Compare medians (center lines), box heights (IQR/spread), whiskers, and any outliers.


```python
plt.figure()
plt.boxplot(class_data, labels=["Class A", "Class B", "Class C"])
plt.title("Test Score Distribution by Class")
plt.ylabel("Score")
plt.tight_layout()
plt.show()

# Basic numerical summary for context
def five_number_summary(x):
    q1 = np.percentile(x, 25)
    q2 = np.percentile(x, 50)
    q3 = np.percentile(x, 75)
    return (np.min(x), q1, q2, q3, np.max(x))

for name, x in zip(["Class A", "Class B", "Class C"], class_data):
    print(name, "min, Q1, median, Q3, max:", five_number_summary(x))

print("Interpretation: Class C exhibits higher median scores and relatively tight spread; "
      "Class B is lower overall; Class A is intermediate. Boxplots are appropriate for quick distribution comparisons.")
```

**Box-and-Whisker Plot**

This code makes a box-and-whisker plot.

* It displays the distribution of test scores for three classes simultaneously.

* Each box highlights the median and the IQR, while the whiskers stretch out to cover the data spread.
* This makes it easy to compare how the scores are spread out and where the center is for each group.

* By running a quick five-number summary (min, Q1, median, Q3, max), I can interpret the results more precisely.
